```python
import pandas as pd
```

```python
teparyDf = pd.read_csv('./../results/Tepary_means.csv')
teparyDf.head()
```

```python
envs = ['M3-AMBIENTE', 'INV-M3-1', 'INV-M3-2', 'Caribia', 'Espinal']
responeses =['YdHa','YdPl','PHI', 'SW', 'PSN', 'SWP','HI','VPP', 'StWP']
teparyRenamedDf= list()
for i in envs:
    envDf = teparyDf[teparyDf['Location'] == i].copy()
    
    if i == 'M3-AMBIENTE':
        rename_cols = {response: '%s_%s' % (response, 'M3-Ambiente') for response in responeses}
    else:
        rename_cols = {response: '%s_%s' % (response, i) for response in responeses}
    envDf.rename(columns=rename_cols, inplace=True)
    envDf.index = envDf.Genotype
    teparyRenamedDf.append(envDf)
    
Tepary_MERGED = pd.concat(teparyRenamedDf, axis = 1)
```

# Merged dataset BLUPs

```python
envs = ['M3-Ambiente', 'INV-M3-1', 'INV-M3-2', 'Caribia', 'Espinal']
data = list()
for i in envs:
    idf = pd.read_excel('./../results/MrBean/BLUPS_GCDT_Merged.xlsx', sheet_name=i,engine = 'openpyxl')
    ipvt = pd.pivot_table(idf, index='Genotype', columns='Trait', values='predicted.values')
    rename_cols = {response: '%s_%s' % (response, i) for response in ipvt.columns}
    ipvt.rename(columns=rename_cols, inplace=True)
    data.append(ipvt)
    print(i)
    print(ipvt.columns)
    print('_'*10)
MERGED = pd.concat(data, axis=1)
MERGED.to_csv('./../results/merged_GCDT_blups_GWAS.csv', na_rep='NA')
#Tepary Parentals means
teparyMered = pd.concat([MERGED , Tepary_MERGED[MERGED.columns]])
MERGED.to_csv('./../results/merged_GCDT_blups_Tepary_GWAS.csv', na_rep='NA')
```
